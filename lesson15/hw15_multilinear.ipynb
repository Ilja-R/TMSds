```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

# Simple linear regression with sklearn

```python
X, y = make_regression(n_samples=100, n_features=1, noise=10)

plt.scatter(X, y)
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
a, b = model.coef_[0], model.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, a * X + b, color='red', label=f'Linear regression = {round(a, 2)}x + {round(b, 2)}')
plt.legend(prop={'size': 10})
plt.show()
```

# Multiple linear regression

```python
X, y = make_regression(n_samples=100, n_features=4, n_informative=4, noise=10, random_state=42)

display(X[:5], y[:5])
```

```python
model = LinearRegression()
model.fit(X, y)
model.coef_, model.intercept_
```

```python
y_pred = model.predict(X)
```

```python
print(y_pred[:5])
```

```python
df = pd.DataFrame(X)
df['y'] = y
df['y_pred'] = y_pred
df.head()
```

```python
df['y_diff'] = df['y'] - df['y_pred']
```

```python
np.mean(df['y_diff'] ** 2)
```
