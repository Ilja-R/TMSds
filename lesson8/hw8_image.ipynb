```python
import matplotlib.pyplot as plt

img = plt.imread("anthracnose1004_.jpg")

plt.imshow(img)
```

```python
data = img.copy()
grayscale_img = data[:, :, 1]
plt.imshow(grayscale_img, cmap ='hot')
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Highlighted Anomalies')
plt.imshow(grayscale_img, cmap ='hot')
plt.axis('off')

plt.show()
```
